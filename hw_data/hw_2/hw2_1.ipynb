Частина перша

- Вивести перші рядки таблиці за допомогою методу head

```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline


url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8"
table = pd.read_html(url, match="Коефіцієнт народжуваності [у|в] регіонах України .*")
df = pd.DataFrame(table[0])
df.head()
```

- Визначте кількість рядків та стовпців у датафреймі (атрибут shape)

```python
rows, columns = df.shape

print(f"Рядов: {rows}")
print(f"Столбцов: {columns}")
```

 - Замініть у таблиці значення "—" на значення NaN

```python
df = df.replace('—', 'NaN')
print(df)
```

- Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
df.dtypes
```

- Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
df['2014'] = df['2014'].astype('float64')
df['2019'] = df['2019'].astype('float64')
print(df.dtypes)
```

- Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
(df.isnull().sum() / len(df)) * 100
```

- Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
df.drop(df.index[-1], inplace=True)
df.tail()
```

- Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
df[['1950', '1960', '1970', '2014', '2019']] = df[['1950', '1960', '1970', '2014', '2019']]. fillna (df[['1950', '1960', '1970', '2014', '2019']].mean())
df = df.round(2)
print(df)
```

- Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
year = '2019'
column = 'Регіон'
condition = df[year] > df.mean(numeric_only=True)[year]
df[condition][column].tolist()
```

- У якому регіоні була найвища народжуваність у 2014 році?

```python
year = "2014"
df.loc[df[year].idxmax(), column]
```

- Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
plt.bar(df['Регіон'], df['2019'], width=0.9)
plt.title('Коефіцієнт народжуваності в 2019 році')
plt.tick_params(labelrotation = 90)
plt.show()
```

- Лінійний графік коефіцієнта народжуваності в Киеві з 1950 по 2019

```python
import matplotlib.pyplot as plt

labels = ['1950', '1960', '1970', '1990', '2000', '2012', '2014', '2019']
fig, ax = plt.subplots(figsize=(10, 10))
kyiv_data = df[df['Регіон'] == 'Київ']
plt.plot(labels, kyiv_data.iloc[0, 1:], color="r")
plt.xticks(labels, rotation=90)
plt.xlabel("Рік", fontsize="small", color="g")
plt.ylabel("Коефіцієнт народжуваності", fontsize="small", color="g")
plt.title("Коефіцієнт народжуваності в Києві з 1950 по 2019", fontsize=15, color="r")
plt.show()
```

- Коефіцієнт народжуваності у регіонах України у 2019

```python
fig, ax = plt.subplots(figsize=(8,8))
plt.plot(df['Регіон'], df['1950'], color="r")
plt.xticks(rotation=90)
plt.xlabel("Коефіцієнт народжуваності", fontsize="small", color="r")
plt.ylabel("Регіон", fontsize="small", color="r")
plt.title("Коефіцієнт народжуваності у регіонах України у 2019", fontsize=15, color="r")
plt.show()
```

- Коефіцієнт народжуваності у регіонах України у 1950 та 2019

```python
fig, ax = plt.subplots(figsize=(8,8))
plt.plot(df['Регіон'], df['1950'], color="r", linewidth=2, marker="o", label="1950")
plt.plot(df['Регіон'], df['2019'], color="b", linewidth=2, marker="v",label="2019")
plt.xticks(rotation=90)
plt.xlabel("Коефіцієнт народжуваності", fontsize=12, color="g")
plt.ylabel("Регіон", fontsize=12, color="g")
plt.title("Коефіцієнт народжуваності у регіонах України у 1950 та 2019", fontsize=15, color="r")
plt.legend()
plt.show()
```
